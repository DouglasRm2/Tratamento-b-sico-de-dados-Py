```
import pandas as pd
import seaborn as sns
```

```
vendas = pd.read_excel("varejo.xlsx")
vendas.head()
```

```
vendas.info()
```

```
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```
vendas.groupby("bandeira").idcompra.nunique()
```

```
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```
vendas.groupby("estado").idcompra.nunique()
```

```
vendas["Nome_Departamento"] = vendas ["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```
vendas.info()
```

```
vendas[vendas["estado"].isnull()]
```

```
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```
vendas["estado"].fillna("MS", inplace=True)
vendas.info()
```

```
media_preco = vendas["Preço"].mean()
vendas["Preço"].fillna(media_preco, inplace=True)
vendas.info()
```

```
vendas.describe()
```

```
vendas[["Preço", "Preço_com_frete"]].describe()
```

```
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_corretas = vendas.query("Preço < Preço_com_frete")
```

```
vendas_corretas.describe()
```

```
vendas_corretas.head()
```

```
vendas_corretas.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```
vendas_corretas.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```
round(vendas_corretas.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```
vendas_corretas.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```
vendas_corretas["mes"] = vendas_corretas["Data"].dt.month
vendas_corretas.head()
```

```
vendas_corretas.groupby('mes').idcompra.nunique().sort_values(ascending=False)
```

```
varejoCliente = pd.read_excel("/content/cliente_varejo.xlsx")
varejoCliente.head()
```

```
vendas_corretas.head()
```

```
varejoCliente[["idade", "renda"]].describe()
```

```
varejoCliente.info()
```

```
varejoCliente = varejoCliente.astype({"renda":"float"})
varejoCliente.info()
```

```
vendas_c = vendas_corretas.merge(varejoCliente, how="left", on="cliente_Log")
vendas_c.head()
```

```
canal_de_venda = round(vendas_c.groupby("idcanalvenda" )["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
canal_de_venda.head()
```

```
agg_idade_bandeira = round(vendas_c.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira.head()
```

```
import matplotlib.pyplot as plt
```

```
fig = plt.figure(figsize=(6,5))

plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="green")
plt.ylabel("media de idade")
plt.title("idade média por Bandeira", loc="center", fontsize=18, fontweight="bold", color="black");
```

```
fig = plt.figure(figsize=( 6,5))
plt.bar(canal_de_venda["idcanalvenda"],canal_de_venda["renda"], color="red")
plt.ylabel("renda média ")

plt.title("renda média por canal de venda", loc="center",fontsize=18, fontweight="bold", color="blue");
```

```
vendas_por_data = vendas_corretas.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data.head()
```

```
fig = plt.figure(figsize=(20,4))

plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = vendas_por_data["Data"]
y = vendas_por_data["idcompra"]

plt.plot(x,y, color="purple", marker="2", linewidth = "2" )
plt.show()
```

```
agg_dep_preco = round(vendas_corretas.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_dep_preco.head()
```

```
fig = plt.figure(figsize=( 20,20))
plt.bar(agg_dep_preco["Nome_Departamento"],agg_dep_preco["Preço_com_frete"], color="red")
plt.ylabel("preço com frete ")

plt.title("Preço médio por departameto", loc="center",fontsize=18, fontweight="bold", color="blue");
```
